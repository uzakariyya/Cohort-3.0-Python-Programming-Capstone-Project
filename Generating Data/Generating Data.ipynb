```python
first_five = [x**3 for x in range(1, 6)]

#This is the Size of the graph which 10 by 6
plt.figure(figsize=(10, 6))

plt.plot(range(1, 6), first_five, marker='<', label='First 5 Cubic Numbers') #'<' This is the marker of the graph and the Label

plt.title('First 5 Cubic Numbers') #This the Title of the graph

#Belwo is the x and y label
plt.xlabel('Numbers')
plt.ylabel('Cubes')

#This is the Legend and Grid of the graph
plt.legend()
plt.grid()
plt.show()
```

```python
first_5000 = [i**3 for i in range(1,5001)]
first_5000

#This is the Size of the graph which 12 by 7
plt.figure(figsize=(12, 7))

plt.plot(range(1, 5001), first_5000, marker='', label="First 5000 Cubic Numbers") #This is the graph and the Label the marker is absence due to the dense of the number


plt.title("First 5000 Cubic Numbers") #This the Title of the graph

#Belwo is the x and y label which named as Numbers and Cubes respectively 
plt.xlabel("Numbers")

plt.ylabel("Cubes")

#This is the Legend and Grid of the graph
plt.legend()

plt.grid()

plt.show()

```

```python
plt.figure(figsize=(12, 7))

# This is the color graient which start from 0 - 13 = 5000
colors = np.linspace(0, 13, 5000)  

plt.scatter(range(1, 5001), first_5000, c=colors, cmap='tab20c', s=3) #This 's' represent the thickness of the line as 3 inches

plt.colorbar(label="Gradient Scale")

plt.title("First 5000 Cubic Numbers with Colormap")

plt.xlabel("Numbers")
plt.ylabel("Cubes")
plt.grid()
plt.show()
```

## Part 2: Random Walk
### TRY IT YOURSELF


```python

#This is the process of defining the Class
class RandomWalk:

# This is the process of initiating and runing an attribute

    def __init__(self, num_points=5000):
        self.num_points = num_points
        self.x_values = [0]
        self.y_values = [0]

#This is the process of creating method which get step
    def get_step(self):
        direction = choice([1, -1])
        distance = choice([0, 1, 2, 3, 4, 5, 6, 7, 8])
        return direction * distance

    def fill_walk(self):
        while len(self.x_values) < self.num_points:
            x_step = self.get_step()
            y_step = self.get_step()

            if x_step == 0 and y_step == 0:
                continue

            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step

            self.x_values.append(x)
            self.y_values.append(y)

rw = RandomWalk()
rw.fill_walk()

plt.style.use('classic')
fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(rw.x_values, rw.y_values, linewidth=0.5) 

plt.show()
```
